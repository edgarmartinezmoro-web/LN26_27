## Trigonométricas

```python
import numpy as np
import matplotlib.pyplot as plt

# Seno y Coseno
x = np.arange(0, 3*np.pi, 0.1)
y1 = np.sin(x)
y2 = np.cos(x)

# Gráfica
plt.plot(x, y1, color='green')
plt.plot(x, y2, color='darkblue')
plt.show()
```

```python
x = np.linspace(-2 * np.pi, 2 * np.pi, 1000)
plt.plot(x, np.tan(x))
plt.ylim(-5, 5)
```

### Razones trigonométricas más comunes: Escuadra y cartabón

$$\frac{\pi}{3},\qquad \frac{\pi}{2},\qquad \frac{2\pi}{3} $$

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array([0, 0, 2, 0])
y = np.array([0, 2, 0, 0])

plt.figure(figsize=(6,6))
plt.plot(x, y, marker='o') # 'o' añade puntos en los vértices
plt.gca().set_aspect('equal', adjustable='box') # Proporción 1:1
plt.title("Escuadra")
plt.grid(True)
plt.show()
```

```python
##### Escuadra
import sympy as smp
smp.init_printing()
x,y,h=smp.symbols('x y h')
eq1 = smp.Eq(x**2 + y**2, h**2)
eq2 = smp.Eq(x - y, 0)
solucion = smp.nonlinsolve([eq1, eq2], [x, y])
solucion
```

```python
h = np.sqrt(3)/2 # Altura
p1 = [0, h]
p2 = [-0.5, 0]
p3 = [0.5, 0]

x = [p1[0], p2[0], p3[0], p1[0]]
y = [p1[1], p2[1], p3[1], p1[1]]
z = [p1[0], 0]
t = [p1[1], 0]

plt.figure(figsize=(6,6))
plt.plot(x, y, marker='o') # 'o' añade puntos en los vértices
plt.plot(z, t, marker='o')
plt.gca().set_aspect('equal', adjustable='box') # Proporción 1:1
plt.title("Cartabón")
plt.grid(True)
plt.show()
```

```python
##### Cartabón
import sympy as smp
smp.init_printing()
x, h, h1, h2=smp.symbols('x h h1 h2')
eq1 = smp.Eq(2*x**2, (2 * h2)**2)
eq2 = smp.Eq(h2**2 +  h1**2, x**2)
solucion = smp.nonlinsolve([eq1, eq2], [h2, h1])
solucion
```

## Exponenciales

```python
import matplotlib.pyplot as plt
import numpy as np

# 1. Generar datos (rango de -5 a 5 con 100 puntos)
x = np.linspace(-5, 5, 100)
# 2. Definir función exponencial:
y = 2**x 
y2=2.2**x
y3=2.4**x
# 3. Crear la gráfica
plt.figure(figsize=(8,5))
plt.plot(x, y, label='f(x) = $2^x$', color='blue')
plt.plot(x, y2, label='f(x) = $2.2^x$', color='red')
plt.plot(x, y3, label='f(x) = $2.4^x$', color='green')



# 4. Personalizar
plt.title('Funciones Exponencial')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.axhline(0, color='black',linewidth=1) # Eje x
plt.axvline(0, color='black',linewidth=1) # Eje y
plt.legend()

# 5. Mostrar
plt.show()
```
