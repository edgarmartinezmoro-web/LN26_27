### Límites

$$
\lim_{x\longrightarrow \pi} sin\left(\frac{x}{2}+sin(x)\right)
$$

```python
smp.limit(smp.sin(x/2+smp.sin(x)),x,smp.pi)
```

```python
smp.limit(2*smp.exp(1/x)/(smp.exp(1/x)+1),x,0,dir='+')
```

```python
smp.limit(2*smp.exp(1/x)/(smp.exp(1/x)+1),x,0,dir='-')
```

```python
smp.limit((smp.cos(x)-1)/x,x, smp.oo)
```

### Derivadas

$$
\frac{d}{dx}\left(\frac{y+sin(x)}{1-cos(x)}\right)^2
$$

```python
smp.diff(((y+smp.sin(x))/(1-smp.cos(x)))**2,x)
```

$$
\frac{d}{dx}(f(x+g(x)))
$$

```python
f,g=smp.symbols('f g',cls=smp.Function)
g=g(x)
f=f(x+g)
```

```python
f
```

```python
smp.diff(f,x)
```
